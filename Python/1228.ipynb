```python
#merge 메소드를 이용해서 데이터 연결 : join
person = pd.read_csv('https://raw.githubusercontent.com/SoongMoo/soldesk20231218/main/data/survey_person.csv')
site = pd.read_csv('https://raw.githubusercontent.com/SoongMoo/soldesk20231218/main/data/survey_site.csv')
visited = pd.read_csv('https://raw.githubusercontent.com/SoongMoo/soldesk20231218/main/data/survey_visited.csv')
survey = pd.read_csv('https://raw.githubusercontent.com/SoongMoo/soldesk20231218/main/data/survey_survey.csv')
site, visited, survey, person
```

```python
site, visited
'''
select name, lat, long, ident, site, dated
from site, inner join visited
where
'''
site_visited = site.merge(visited, left_on='name', right_on='site')
site_visited
```

```python
person, survey
'''
select ident, personal, family, taken, person, quant, reading
from person inner join survey
on person.ident = survey.person
'''
person_survey = person.merge(survey, left_on='ident', right_on='person')
person_survey
```

```python
person, survey
'''
select ident, personal, family, taken, person, quant, reading
from person left outer join survey
on person.ident = survey.person
'''
person_survey = person.merge(survey, left_on='ident', right_on='person', how='left')
person_survey #person의 join되지않은 부분을 출력
person_survey = person.merge(survey, left_on='ident', right_on='person', how='right')
person_survey #person의 join되지않은 부분을 출력

person_survey = survey.merge(person, left_on='person', right_on='ident', how='right')
person_survey
person_survey = survey.merge(person, left_on='person', right_on='ident', how='left')
person_survey
```

```python
person_survey = survey.merge(person, left_on='person', right_on='ident')
person_survey = survey.merge(person, left_on='person', right_on='ident', how='inner')
person_survey
```

```python
visited, survey
'''
select ident   site  dated  taken person quant  reading
from visited join survey
on visited.ident = survey.taken
'''
visited_survey = visited.merge(survey, left_on='ident', right_on='taken')
visited_survey
```

```python
visited_survey, person_survey
'''
select  ident   site       dated  taken person quant  reading
        taken person quant  reading ident   personal   family
from visited_survey as v join person_survey as p
on v.ident = p.taken
and v.person = p.ident
and v.quant = p.quant
and v.reading = p.reading
'''
visited_survey_person_survey = visited_survey.merge(person_survey,\
                                                    left_on=['ident', 'person','quant','reading'],\
                                                    right_on=['taken','ident', 'quant', 'reading'])
visited_survey_person_survey
```
